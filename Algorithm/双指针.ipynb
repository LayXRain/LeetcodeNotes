### 移动零问题

```python
#移动零,双指针
def yidong(nums):
    n=len(nums)
    i = 0
    for j in range(n):
        if nums[j]!=0:
            nums[i]=nums[j]
            nums[j]=0
            i+=1
    return nums
nums=[0,1,2,3,5,0,0,225,2452234,530,34,0,678,5]
print(yidong(nums))
```

### 定长滑动窗口

```python
### 滑动窗口问题
#定长滑动窗口
from typing import List
nums:List[int]
m:int
nums = [1,3,5,6,2,-11,12,5,15,23,-1]
'''''长度为m的定长滑动窗口'''
def chaunkou(nums,m):
    left = 0
    n=len(nums)
    #可以初始化窗口，也可以不初始化
    cur_sum =0
    max_sum =float('-inf')
    for i in range(m):
        cur_sum+=nums[i]
    max_sum=cur_sum
    for i in range(m,n):
        cur_sum-=nums[i-m]
        cur_sum+=nums[i]
        max_sum = max(max_sum, cur_sum)
    return max_sum
print(chaunkou(nums,3))
```
