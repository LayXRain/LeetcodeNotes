```python
import bisect
t = int(input())
for _ in range(t):
    n, m = map(int, input().split())
    a = list(map(int, input().split()))
    b = list(map(int, input().split()))
    c = list(map(int, input().split()))
    monsters = sorted(zip(b, c), key=lambda x: (x[0],-x[1]))
    swords = sorted(a)
    ans = 0
    for blood, new_sword in monsters:
        idx = bisect.bisect_left(swords, blood)
        if idx == len(swords):
            break
        cur = swords[idx]
        swords.pop(idx)
        ans += 1
        """这里一起处理是有问题的，先处理有反馈的怪兽，因为保底会获得一个相同伤害量的剑"""
        """"对于c = 0 的怪兽进行相同的处理逻辑是有问题的，因为相对于有反馈的但是血量更大的怪兽使用同一把剑可能收益会更高"""
        if new_sword>0:
            bisect.insort(swords, max(cur,new_sword))
    print(ans)
```

```python
import sys
import bisect
def solve():
    data = sys.stdin.read().strip().split()
    t = int(data[0])
    idx = 1
    results = []
    for _ in range(t):
        n = int(data[idx]); m = int(data[idx+1]); idx += 2
        a = list(map(int, data[idx:idx+n])); idx += n
        b = list(map(int, data[idx:idx+m])); idx += m
        c = list(map(int, data[idx:idx+m])); idx += m
        # 分离怪物
        with_drop = []
        without_drop = []
        for i in range(m):
            if c[i] > 0:
                with_drop.append((b[i], c[i]))
            else:
                without_drop.append(b[i])
        # 关键：有掉落怪物先按c降序，再按b升序
        # 这样我们先处理掉落好的怪物，如果掉落相同，先处理血量低的
        with_drop.sort(key=lambda x: (x[0],-x[1]))
        without_drop.sort()  # 无掉落怪物按血量升序
        # 初始剑排序
        swords = sorted(a)
        ans = 0
        # 处理有掉落怪物
        for blood, drop in with_drop:
            # 找到能打败怪物的剑
            pos = bisect.bisect_left(swords, blood)
            if pos < len(swords):
                used_sword = swords[pos]
                swords.pop(pos)
                ans += 1
                # 获新剑
                new_sword = max(used_sword, drop)
                bisect.insort(swords, new_sword)
        # 处理无掉落怪物
        for blood in without_drop:
            pos = bisect.bisect_left(swords, blood)
            if pos < len(swords):
                swords.pop(pos)
                ans += 1
        results.append(str(ans))
    sys.stdout.write("\n".join(results))
"""逻辑对了但是超时了，我要炸了，同样的算法py超时了，cpp就过了，cpp还真是最适合来写算法的"""
if __name__ == "__main__":
    solve()
```
