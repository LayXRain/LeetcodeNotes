### 一般树

```python
from typing import List, Optional
from collections import deque

#二叉树节点定义
class TreeNode:
    def __init__(self,val,left=None,right=None):
        self.val = val
        self.left = left
        self.right = right


nums = [10,5,-3,3,2,None,11,3,-2,None,1]
# 给定一个数组，按层序遍历来创建一个数组
# 队列先进先出，符合层序遍历从左到右固定处理顺序的逻辑
def CreateBtree(nums:List[int])->Optional[TreeNode]:
    if not nums:
        return None
    root_val = nums[0]
    root = TreeNode(root_val)
    queue = deque([root])
    index= 1
    n = len(nums)
    while queue and index < n:
        cur = queue.popleft()
        left_val = nums[index] if index < n else None
        if left_val is not None:
            cur.left = TreeNode(left_val)
            queue.append(cur.left)
        index +=1
        right_val =nums[index] if index <n else None
        if right_val is not None:
            cur.right = TreeNode(right_val)
            queue.append(cur.right)
        index+=1
    #取出当前层列表
    return root
#按根节点按层序遍历来打印数组，并且空节点添加None
def PrintLevelOrderWithNull(root: Optional[TreeNode]) -> None:
    if not root:
        return
    result = []
    queue = deque([root])  # 队列中可以存TreeNode或None
    while queue:
        # 标记当前层是否有非空节点（用于终止循环，避免无限添加None）
        has_non_null = False
        level_size = len(queue)
        current_level = []
        for _ in range(level_size):
            node = queue.popleft()
            if node is not None:
                # 非空节点，记录值
                current_level.append(node.val)
                # 将左、右子节点加入队列（可能为None）
                queue.append(node.left)
                queue.append(node.right)
                # 标记当前层有非空节点
                if node.left is not None or node.right is not None:
                    has_non_null = True
            else:
                # 空节点，记录None
                current_level.append(None)
                # 空节点的左右子节点也为None，加入队列占位
                queue.append(None)
                queue.append(None)
        # 将当前层结果加入总结果
        result.extend(current_level)
        # 如果当前层之后没有非空节点，终止循环（避免无限添加None）
        if not has_non_null:
            break
    # 处理末尾的全空节点（去掉最后连续的None）
    while result and result[-1] is None:
        result.pop()
    print(result)
root  = CreateBtree(nums)
PrintLevelOrderWithNull(root)
```

```python
def DFS(root: Optional[TreeNode], target=None):
    if target is None:
        target = []
    if not root:
        return target
    # 中序遍历：左-根-右
    if root.left:
        DFS(root.left, target)
    target.append(root.val)
    if root.right:
        DFS(root.right, target)
    return target
target = DFS(root)
print(target)
```

```python
def BFS(root:TreeNode,target=None):#层序遍历
    if target is None:
        target = []
    if not root:
        return []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        target.append(node.val)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return target
target = BFS(root)
print(target)
```
